```python
dataset = [['Milk','Onion','Nutmeg','Kidney Beans','Eggs','Yogurt'],
           ['Dill','Onion','Nutmeg','Kidney Beans','Eggs','Yogurt'],
           ['Milk','Apple','Kidney Beans','Eggs'],
           ['Milk','Unicorn','Corn','Kidney Beans','Yogurt'],
           ['Corn','Onion','Onion','Kidney Beans','Eggs','Ice cream']]

print(dataset)
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

#TransactionEncoder: Returns unique items of the dataset and
#create datatable with binary values
```

```python
#find unique items in dataset
te= TransactionEncoder()
te_ary= te.fit(dataset)
print(te_ary.columns_)
```

```python
pd.DataFrame(te.transform(dataset), columns= te_ary.columns_)
```

```python
#most frequent items: Algorithm apriori

from mlxtend.frequent_patterns import apriori
frequent_itemsets= apriori(pd.DataFrame(te_ary.transform(dataset), columns= te_ary.columns_), min_support=0.4, use_colnames=True)
print (frequent_itemsets)
```

```python
#Calculate Confidence & lift values for most frequent itemsets

from mlxtend.frequent_patterns import association_rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
result= rules[['antecedents','consequents','support','confidence','lift']]
result.head(10)
```

```python
result['antecedents'] = result['antecedents'].astype(str)
result["consequents"] = result["consequents"].astype(str)
```

```python
result['antecedents'].value_counts()
```

```python
rec = result[['antecedents','consequents']]
rec

#rec is recommendation
```

```python
#recommendation for kidney beans where others might also be present on the antecedent list
rec[rec['antecedents'].str.contains('Kidney Beans')]
```

```python
rec['antecedents'] = rec['antecedents'].str.replace('frozenset' , '')
rec['consequents'] = rec['consequents'].str.replace('frozenset' , '')
```
